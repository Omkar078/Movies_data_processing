##### Removing ()


```python
df['Year'].unique()
```

```python
df['Year'] = df['Year'].str.replace(r'\W', '', regex = True)
df['Year'].head()
```

```python
df['Year'].unique()
```

```python
df['Year'] = df['Year'].replace('[a-z[A-Z]', '', regex = True)
```

```python
df['Year'].isnull().sum()
```

```python
df['Year'] = df['Year'].astype(str).apply(lambda x: x[4:] if len(x) > 4 else x)
print(df)
```
